# Example- 2.2(b)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define user-defined functions
def mul(a, b):
    return a * b  # Element-wise multiplication

def shif(x, shift):
    return np.roll(x, shift)  # Shift the signal by 'shift' units

def add(a, b):
    return a + b  # Element-wise addition

# Define the original signal x(n) (e.g., unit impulse or step function)
n = np.arange(-10, 10, 1)  # Discrete time range
x_n = np.zeros_like(n)  # Start with a zero array

# Define x(n) as a unit impulse at n=0
x_n[n == 0] = 1  # Unit impulse at n=0

# Apply the transformation x2(n) = x(3-n) + x(n) * x(n-2)
x_n_shifted_3_minus_n = shif(x_n, -(n - 3))  # x(3 - n)
x_n_shifted_2 = shif(x_n, -2)  # x(n - 2)

# Use the user-defined 'mul' and 'add' functions
x2_n = add(x_n_shifted_3_minus_n, mul(x_n, x_n_shifted_2))

# Plot the original signal and the transformed signal
plt.subplot(2, 1, 1)
plt.stem(n, x_n)
plt.title("Original Signal x(n)")

plt.subplot(2, 1, 2)
plt.stem(n, x2_n)
plt.title("Transformed Signal x2(n) = x(3-n) + x(n) * x(n-2)")

plt.tight_layout()
plt.show()
```

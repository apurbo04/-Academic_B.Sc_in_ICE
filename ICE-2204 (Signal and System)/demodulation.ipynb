```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector

carrier_freq = 50  # Carrier frequency (Hz)
message_freq = 5   # Message signal frequency (Hz)

# Message signal (low-frequency)
message = np.cos(2 * np.pi * message_freq * t)

# Carrier wave (high-frequency)
carrier = np.cos(2 * np.pi * carrier_freq * t)

# Amplitude Modulated (AM) signal
modulated_signal = (1 + 0.5 * message) * carrier

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, modulated_signal, label="AM Signal")
plt.plot(t, message, '--', label="Original Message", alpha=0.7)
plt.legend()
plt.title("Amplitude Modulated (AM) Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
from scipy.signal import hilbert

# Envelope detection using Hilbert Transform
analytic_signal = hilbert(modulated_signal)
envelope = np.abs(analytic_signal)

# Plot the recovered signal
plt.figure(figsize=(10, 4))
plt.plot(t, envelope, label="Recovered Message", color='r')
plt.plot(t, message, '--', label="Original Message", alpha=0.7)
plt.legend()
plt.title("Demodulated Signal Using Envelope Detection")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
